## Se sim, certamente já foi alvo de um teste de hipótese

![fiig](https://miro.medium.com/max/2075/1*3VHfDOniZ-0DJGa1abyCfw.png)

* Null Hypothesis  => média de quem viu um banner é igual do outro
* Alternative Hypothesis  => média de quem viu o banner 1 é diferente do outros

Ou seja:

    H0: m1 == m2
    H1: m1 != m2
    
Importante, qual nível de confiabilidade?

Geralmente usamos 95%

O que equivale a Pvalue = 0.05

Ou seja.
* Se P < 0.05 rejeitamos H0, as médias são significantemente diferentes.
* Se P > 0.05 aceitamos H0, as médias não são significamente diferentes.

Para fazer um teste t:

    from scipy.stats import ttest_ind
    data1, data2 = ...
    stat, p = ttest_ind(data1, data2)
