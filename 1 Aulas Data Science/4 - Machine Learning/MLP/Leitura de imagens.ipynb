# Lao que leia diversos arquivos

```python
import os

folder = r'C:\Users\rodri\OneDrive\Imagens\exemplo imagens categorizadas' #  so colocar aqui a pasta que quer utilizar
```

```python
# Lendo todos arquivos de uma pasta
for file in os.listdir(folder):
    print(file)
```

```python
# Manipulando strings
file.split('.')[0]
```

```python
# Juntando o diretorio com o arquivo
diretorio = r'C:\Users\rodri\OneDrive\Imagens\exemplo imagens categorizadas'
os.path.join(diretorio, file)
```

```python
lista_respostas = []
lista_arrays = []

for file in os.listdir(diretorio):
    lista_respostas.append(file.split('.')[0])
    
    arquivo = os.path.join(diretorio, file)
    im=cv2.imread(arquivo,0)
    
    resized = cv2.resize(im, (300,300))
    
    plt.imshow(resized, cmap='gray');
    plt.axis('off')
    plt.show()
    
    array = resized.flatten()
    
    lista_arrays.append(array)
```
